```python
import random as rd
import matplotlib.pyplot as plt
```

```python
def step(pos, b=False):
    '''
    takes current position pos
    returns next_pos
    '''
    #Special case 1
    if pos == -4:
        next_pos = -3
    
    #If doing part b of excersize 2.
    elif b and pos == 6:
        #Partial block at 6: 0.25 prob of going to 7.
        if rd.uniform(0,1) < 0.25:
            next_pos = 7
        else:
            next_pos = 5
            
    #Normal coinflip
    else:
        next_pos = pos + rd.choice([1, -1])
    
    return next_pos
```

# Plotting a sample path for Part a)

```python
steps = 30
path = [0]
for i in range(steps-1):
    pos = path[-1]
    path.append(step(pos))
```

```python
plt.figure(figsize=(12, 8))
plt.step(list(range(steps)), path)
plt.title("30 flips of a biased coin with P(Heads)=0.6")
plt.xlabel("Step")
plt.ylabel("Distance from starting point")
plt.show()
```

It reaches -4 and then bounces back. Other than that, just a normal random walk.

# Comparing distributions over final position of walk after 100 steps for part a) and b)

```python
def sim(b=False, steps=100):
    #Start at 0
    pos = 0
    for _ in range(steps):
        pos = step(pos, b)
    return pos

final_pos_list_a = [sim() for _ in range(1000)]
final_pos_list_b = [sim(True) for _ in range(1000)]
```

```python
plt.figure(figsize=(14, 10))
plt.hist(final_pos_list_b, alpha=0.5, label="Part a) + partial block at x=6")
plt.hist(final_pos_list_a, alpha=0.5, label="Part a)")
plt.legend(loc=0)
plt.xlabel("Final position")
plt.ylabel("Frequency (out of 100)")
plt.show()
```

We see that the distribution over the final positions for part b (blue) is biased more closely towards zero. This makes sense given the reduced probability os going past x=6 implied by the rules. 
Once passed x=6, however, the rules of the two simulations are identical, meaning that the probability of going further or returning to below 6 is unaffected.

```python
def step(x, y, sigma):
    '''
    Takes position, x y
    Sigma, sd of step size
    returns next position
    '''
    x_direction = np.random.uniform(-1, 1)
    y_direction = np.random.uniform(-1, 1)
    step_size = np.random.normal(0, sigma)

    new_x = x + (x_direction*step_size)
    new_y = y + (y_direction*step_size)
    return new_x, new_y

def sim(steps, sigma):
    x = 0
    y = 0
    for s in range(steps):
        x, y = step(x, y, sigma)
    return x, y

results = {}
for sigma in [0.5, 1, 2]:
    for steps in [20, 400, 1600]:
        results["%d"%sigma]["%d"%steps] = [sim(steps, sigma) for _ in range(1000)]
```
