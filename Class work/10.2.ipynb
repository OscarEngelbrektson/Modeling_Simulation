```python
money_at_death=[]

for i in range(1000):
    savings = 150000
    for year in range(50, 106):

        #Assume she invests and withdraws at beginnning of year, 
        #so that it affects interest earned that year
        if year in list(range(50, 71)):
            savings += 10000
        #if older than 70, she withdraws
        else:
            savings -= 65000

        # new savings given interest
        savings *= sts.norm(0.08, 0.09).rvs()

        #Does she die?
        #if so, record savings at death and break the loop
        if 1-data[year] < rn.random():
            money_at_death.append(savings)
            break
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.hist(money_at_death)
plt.title("Fannie Mae: savings at death")
plt.xlabel("Savings at death")
plt.ylabel("Frequency (Times in 1000)")
plt.show()
```

The mode indicates that the most likely scenario, is that Fannie Mae will die with a low, positive savings balance. This is the ideal outcome, as she will have lived maximally on the savings she has accrued in her life.
However, there is a  significant probability (all probability mass below the 0 on the x-axis, 57%) she will have used up all her savings before she dies. Conversely, there is a low probability of her having excess savings at time of death.

Realistically, one might want to play it safe and live of less than 65000 per year in retirement so as to decrease the probability of running out of money before death.
